```python
import pandas as pd
from pandas.plotting import radviz
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.io.pickle import to_pickle
from collections import Counter
from imblearn.under_sampling import RandomUnderSampler, TomekLinks
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.combine import SMOTEENN, SMOTETomek
from imblearn.ensemble import BalancedBaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV, KFold, train_test_split
from sklearn.preprocessing import StandardScaler
import time
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, IsolationForest
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC

plt.style.use('fivethirtyeight')

%matplotlib inline
pd.options.display.max_columns = 400
pd.options.display.max_rows = 5000
```

```python
df = pd.read_pickle('datasets/cleaned.pkl')
```

```python
labels = pd.read_pickle('datasets/labels.pkl')
```

```python
features = ['Protocol_6','PSH_Flag_Count','Bwd_Packet_Length_Std','Fwd_IAT_Std','min_seg_size_forward','Bwd_Packets_persec',
            'FIN_Flag_Count','Idle_Std','Flow_IAT_Mean','Bwd_IAT_Std','Bwd_IAT_Total','Init_Win_bytes_backward',
            'SYN_Flag_Count','Avg_Fwd_Segment_Size','Fwd_Packet_Length_Min','URG_Flag_Count','Destination_Port']
```

```python
y = df.loc[:, 'Label']
X = df.loc[:, features]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.25, random_state=101, stratify=y)
```

```python
labellist = [l for label in labels.Label.values for l in label]
labellist.sort()
labellist
```

```python
ss = StandardScaler()
Xs_train = ss.fit_transform(X_train)
Xs_test = ss.fit_transform(X_test)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(Xs_train)

# Percentage of variance explained for each components
print('explained variance ratio (first two components): %s'
      % str(pca.explained_variance_ratio_))

plt.figure(figsize=(20,10)); colors=['#3cb44b','#ffe119','#0082c8','#f58231','#911eb4','#46f0f0','#f032e6','#d2f53c',
                                   '#008080','#e6beff','#aa6e28','#fffac8','#800000','#000080','#e6194b']; lw = 2

for color, i, label_name in zip(colors, labellist, labellist):
    plt.scatter(X_pca[y_train == i, 0], X_pca[y_train == i, 1], c=color, alpha=.8, lw=lw, label=label_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of "Label"')
plt.tight_layout(); plt.margins(0.02)
plt.show()
```

```python
'''RadViz is a multivariate data visualization algorithm that 
plots each axis uniformly around the circumference of a circle 
then plots points on the interior of the circle such that the 
point normalizes its values on the axes from the center to 
each arc'''

colors=['#3cb44b','#ffe119','#0082c8','#f58231','#911eb4','#46f0f0','#f032e6','#d2f53c',
        '#008080','#e6beff','#aa6e28','#fffac8','#800000','#000080','#e6194b']


plt.figure(figsize=(20,10))
radviz(df[features + ['Label']], 'Label', color=colors)
plt.show()
```

## Initial Model

Here we will begin with running Logistic Regression without handling the imbalance in data, this will be the benchmark result that we will aim to better. 

The instinct is that without handling the heavily imbalanced data the trained model will have difficulties predicting the minority classes on the test set. On top of that, as seen from the visualisation, the classes seem to overlap each other quite badly. A linear classifier might not perform too well in classifying this dataset.

Nevertheless, we will still use logistic regression as the benchmark classifier to compare results.

```python
# Logistic Regression
logreg = LogisticRegression(solver='sag', multi_class='multinomial', n_jobs=3, verbose=1)
logreg.fit(Xs_train, y_train)

print('Accuracy Score: ' + str(logreg.score(Xs_test, y_test)))
print('=============================================================')
print('Classification Report:')
print(classification_report(y_test, logreg.predict(Xs_test)))
print('=============================================================')
print('Confusion Matrix:')
```

```python
pd.DataFrame(confusion_matrix(y_test, logreg.predict(Xs_test)), index=labellist, columns=labellist)
```

As expected, the model has difficulty predicting the minority classes, with most of them having 0 recall and precision. 

The number of hostile traffics that were misclassified as benign traffic seems large.

We will now proceed with a simple balancing of data by choosing balanced for the logistic regression's class weights and see how much the result improve.

```python
# Logistic Regression with balanced class weights
logreg = LogisticRegression(class_weight='balanced', solver='sag', multi_class='multinomial', n_jobs=3, verbose=1)
logreg.fit(Xs_train, y_train)

print('Accuracy Score: ' + str(logreg.score(Xs_test, y_test)))
print('=============================================================')
print('Classification Report:')
print(classification_report(y_test, logreg.predict(Xs_test)))
print('=============================================================')
print('Confusion Matrix:')
```

```python
pd.DataFrame(confusion_matrix(y_test, logreg.predict(Xs_test)), index=labellist, columns=labellist)
# 412
```

This is quite surprising, by just assigning more weights to the minority classes when performing logistic regression actually improves the result quite significantly.

However, there still seem to be quite a number of mis-classification.

We will now proceed to try out some resampling methods using the imbalanced-learn package.

## Random undersampling

For random undersampling, by default the ratio to be resampled is set as 'auto' where the all classes are shrunk to the count of the smallest minority class. 

In this case the other classes were shrunk to 8 counts. The logistic regression results was much worse than the benchmark result.

Therefore, an ideal resampled ratio has to be entered to prevent that. Here, the classes with counts > 4000 were shrunk to 4000 and those with < 4000 will remain as they are. 

```python
ratio = {'BENIGN': 4000, 'DoS Hulk': 4000, 'PortScan': 4000, 'DDoS': 4000, 'DoS GoldenEye': 4000, 'FTP-Patator': 4000,
         'SSH-Patator': 4000, 'DoS slowloris': 4000, 'DoS Slowhttptest': 4000}

rus = RandomUnderSampler(ratio=ratio, random_state=101)
X_train_rus, y_train_rus = rus.fit_sample(Xs_train, y_train)

print("Distribution of class labels before resampling {}".format(Counter(y_train)))
print("Distribution of class labels after resampling {}".format(Counter(y_train_rus)))
```

```python
# pd.to_pickle(X_toytrain_rus, 'datasets/X_toytrain_rus.pkl')
# pd.to_pickle(y_toytrain_rus, 'datasets/y_toytrain_rus.pkl')
```

```python
pca2 = PCA(n_components=2)
Xrs_pca = pca2.fit_transform(X_train_rus)

# Percentage of variance explained for each components
print('explained variance ratio (first two components): %s'
      % str(pca2.explained_variance_ratio_))

plt.figure(figsize=(20,10)); colors=['#3cb44b','#ffe119','#0082c8','#f58231','#911eb4','#46f0f0','#f032e6','#d2f53c',
                                   '#008080','#e6beff','#aa6e28','#fffac8','#800000','#000080','#e6194b']; lw = 2

for color, i, label_name in zip(colors, labellist, labellist):
    plt.subplot(2,2,1)
    plt.scatter(Xrs_pca[y_train_rus == i, 0], Xrs_pca[y_train_rus == i, 1], c=color, alpha=.8, lw=lw, label=label_name)
    plt.legend(loc='best', shadow=False, scatterpoints=1)
    plt.title('PCA of "Label"')
    plt.tight_layout(); plt.margins(0.02)
    
    plt.subplot(2,2,2)
    plt.scatter(X_pca[y_train == i, 0], X_pca[y_train == i, 1], c=color, alpha=.8, lw=lw, label=label_name)
    plt.legend(loc='best', shadow=False, scatterpoints=1)
    plt.title('PCA of "Label"')
    plt.tight_layout(); plt.margins(0.02)

plt.show()
```

```python
# Logistic Regression with l2 penalty
rusclf = LogisticRegression(solver='sag', multi_class='multinomial', n_jobs=-1, verbose=1, max_iter=100000)

rusclf.fit(X_train_rus, y_train_rus)

print('Accuracy Score: ' + str(rusclf.score(Xs_test, y_test)))
print('=============================================================')
print('Classification Report:')
print(classification_report(y_test, rusclf.predict(Xs_test)))
print('=============================================================')
```

```python
pd.DataFrame(confusion_matrix(y_test, rusclf.predict(Xs_test)), index=labellist, columns=labellist)
#198
```

Results seem rather decent with only 198 attacks classified as benign traffic, 

However the hostile classes, 'Web Attack – Sql Injection' and 'Web Attack – XSS' have 0 True Positives.


Next, we will try if regularization with 'l1' penalty will improve the results. 

A gridsearch was ran on prior tries and a C of 1.0(default) gave the best results, which also meant no regularization was applied.

The results seem similar, but has 218 attacks classified as benign traffic.

```python
print(rusclf.best_params_)
print(rusclf.best_score_)
```

```python
# Logistic Regression with l1 penalty
rusclf = LogisticRegression(penalty='l1', solver='saga', multi_class='multinomial', n_jobs=-1, verbose=1, max_iter=100000)

rusclf.fit(X_train_rus, y_train_rus)

print('Accuracy Score: ' + str(rusclf.score(Xs_test, y_test)))
print('=============================================================')
print('Classification Report:')
print(classification_report(y_test, rusclf.predict(Xs_test)))
print('=============================================================')
```

```python
pd.DataFrame(confusion_matrix(y_test, rusclf.predict(Xs_test)), index=labellist, columns=labellist)
# 218
```

```python
# Logistic Regression with balanced class weights
rusclf = LogisticRegression(solver='sag', class_weight='balanced', multi_class='multinomial', 
                            n_jobs=-1, verbose=1, max_iter=100000)

rusclf.fit(X_train_rus, y_train_rus)

print('Accuracy Score: ' + str(rusclf.score(Xs_test, y_test)))
print('=============================================================')
print('Classification Report:')
print(classification_report(y_test, rusclf.predict(Xs_test)))
print('=============================================================')
```

```python
pd.DataFrame(confusion_matrix(y_test, rusclf.predict(Xs_test)), index=labellist, columns=labellist)
#152
```

Lastly, a balanced class weight was applied to give more weightage to the smaller classes.

Overall results seems better, with even lesser attacks(152) classified as benign traffic

'Web Attack – Sql Injection' and 'Web Attack – XSS' did not go undetected this time round.

### Customised Under and Over sampling: Multi-resampler

As the previous sampling methods only managed to either down sample or up sample, I thought it might be a good idea to combine both methods.

The following function first finds the mean or median count of all classes. Next, it under sample the classes with counts higher than mean/median to that mean/median count. Lastly, it over sample the classes with counts lower than mean/median to that mean/median count.

With this method the class count will balance up nicely.

```python
def multi_resampler(X, y, undersampler=RandomUnderSampler(), oversampler=RandomOverSampler(), ratio='mean', random_state=None):
    '''
    Finds the mean count of labels or the median count of labels.
    Undersample classes with counts higher than mean/median to mean/median count.
    Then, oversample classes with counts lower than mean/median to mean/median count.
    
    Parameters:
    X : {array-like, sparse matrix}, shape (n_samples, n_features) Matrix containing the data which have to be sampled.

    y : array-like, shape (n_samples,) Corresponding label for each sample in X.
    
    undersampler : underlying sampler class/function to perform undersampling
    
    oversampler : underlying sampler class/function to perform oversampling
    '''
    if ratio == 'mean':
        mean = int(round(np.mean(list(Counter(y_train).values()))))
        to_undersample = {key: mean for key, value in Counter(y).items() if value >= mean}
        to_oversample = {key: mean for key, value in Counter(y).items() if value < mean}
        
    elif ratio == 'median':
        median = int(np.median(list(Counter(y_train).values())))
        to_undersample = {key:median for key, value in Counter(y).items() if value >= median}
        to_oversample = {key:median for key, value in Counter(y).items() if value < median}

    us = undersampler
    us.ratio = to_undersample
    us.random_state = random_state
    X_us, y_us = us.fit_sample(X, y)
    
    os = oversampler
    os.ratio = to_oversample
    os.random_state=random_state
    X_os, y_os = os.fit_sample(X_us, y_us)
    
    print("Distribution of class labels before resampling {}".format(Counter(y)))
    print('Distribution of class labels after resampling {}'.format(Counter(y_os)))
    return X_os, y_os
```

I first tried using the mean count as the ratio to balance the data up. The under sampler and over sampler used were set as default (RandomUnderSampler and RandomOverSampler).

```python
X_ms, y_ms = multi_resampler(Xs_train, y_train)
```

```python
pca2 = PCA(n_components=2)
Xrs_pca = pca2.fit_transform(X_ms)

# Percentage of variance explained for each components
print('explained variance ratio (first two components): %s'
      % str(pca2.explained_variance_ratio_))

plt.figure(figsize=(20,10)); colors=['#3cb44b','#ffe119','#0082c8','#f58231','#911eb4','#46f0f0','#f032e6','#d2f53c',
                                   '#008080','#e6beff','#aa6e28','#fffac8','#800000','#000080','#e6194b']; lw = 2

for color, i, label_name in zip(colors, labellist, labellist):
    plt.subplot(2,2,1)
    plt.scatter(Xrs_pca[y_ms == i, 0], Xrs_pca[y_ms == i, 1], c=color, alpha=.8, lw=lw, label=label_name)
    plt.legend(loc='best', shadow=False, scatterpoints=1)
    plt.title('PCA of "Label"')
    plt.tight_layout(); plt.margins(0.02)
    
    plt.subplot(2,2,2)
    plt.scatter(X_pca[y_train == i, 0], X_pca[y_train == i, 1], c=color, alpha=.8, lw=lw, label=label_name)
    plt.legend(loc='best', shadow=False, scatterpoints=1)
    plt.title('PCA of "Label"')
    plt.tight_layout(); plt.margins(0.02)

plt.show()
```

## Conclusion
To conclude, the customised resampler seems to work best. By finding a mid point (the median), it balances the data without losing too much information from each class.
With regards to the modelling, Logistic Regression is still the most effective, only 71 which is only 0.06% hostile traffic flows went undetected.
However, there is a tradeoff in the amount of false alarms there are (in this case 193,204 around 33% of all benign flows in the test set). 
If the model is deployed and solely relied on, the cost of looking into these false alarms could be high. 
That, historically, has been the issue of anomaly-based intrusion detection systems.
